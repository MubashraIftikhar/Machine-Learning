### Task 21
Add a new column Age Group with values "Young", "Middle-aged", or "Senior" based on Age.

```python
def age_group(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 60:
        return 'Middle-aged'
    else:
        return 'Senior'

if 'Age Group' not in df.columns:
    df['Age Group'] = df['Age'].apply(age_group)
else:
    df.loc[:, 'Age Group'] = df['Age'].apply(age_group)

print("\nDataFrame after adding the 'Age Group' column:")
df.head()
```

### Task 48
Create a DataFrame with employee ID and Project Count, then merge it with the original dataset. 
Calculate the average Salary per project.

```python
project_count = {'ID': [1, 2, 3, 4, 5],'Project Count': [3, 2, 4, 5, 1]}
project_count_df = pd.DataFrame(project_count)

# Merge the dataset with project count information
df_with_project_count = df.merge(project_count_df, on='ID', how='left')

# Calculate the average Salary per project
df_with_project_count['Average Salary per Project'] = df_with_project_count['Salary'] / df_with_project_count['Project Count']
print(df_with_project_count)
```

### Task 49
Calculate the percentage increase in Salary for each employee compared to the average Salary in
their Department.

```python
# Calculate the average Salary in each Department
average_salary_by_department = df_with_project_count.groupby('Department')['Salary'].transform('mean')

# Calculate the percentage increase in Salary for each employee
df_with_project_count['Percentage Increase in Salary'] = (
    (df_with_project_count['Salary'] - average_salary_by_department) / average_salary_by_department) * 100
print(df_with_project_count)
```
